# Outliers Removal

```python
#Removing outliers using zscore
from scipy.stats import zscore
z=np.abs(zscore(df))
threshold=3
new_df=df[(z<3).all(axis=1)]
```

```python
#Removing skewness using log method
skew=('Total Volume','4046','4225','4770','Total Bags','Small Bags','Large Bags','XLarge Bags','year')
for col in skew :
    if new_df.skew().loc[col]>0.55:
        new_df[col] = np.log1p(new_df[col])
```

# Scaling

```python
#Scaling data using Standard Scaler
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
ScaledX=sc.fit_transform(X_reg)
```

#  Regression Models

```python
#importing libraries for models
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error 

best_rstate=0
accu=0
best_model=[]
mod=[LinearRegression(),DecisionTreeRegressor(),KNeighborsRegressor(),RandomForestRegressor(),GradientBoostingRegressor(),SVR()]
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(ScaledX,Y_reg,test_size=0.33,random_state=i)
    for m in mod:
        m.fit(x_train,y_train.values.ravel())
        y_pred=m.predict(x_test)
        tempaccu=r2_score(y_test,y_pred)
        if tempaccu>accu:
            accu=tempaccu
            best_rstate=i
            best_model=m
            print(f"best accuracy : {accu*100} found on the random state {best_rstate}, best model is {m}")
            
print(f"-----best accuracy : {accu*100} found on the random state {best_rstate}, best model is {m}----")
```

# Hyperparameter Tuning

```python
#Grid Search CV for Random forest regressor
from sklearn.model_selection import GridSearchCV

x_train,x_test,y_train,y_test=train_test_split(ScaledX,Y_reg,test_size=0.33,random_state=123)
RFR=RandomForestRegressor()
parameters = { 
            "n_estimators" : [10,20,30],
            "max_features" : ["auto", "sqrt", "log2"],
            "min_samples_split" : [2,4,8],
            "bootstrap" : [True, False],
            }
GCV=GridSearchCV(estimator=RFR,param_grid=parameters,cv=3,n_jobs=-1)
GCV.fit(x_train,y_train.values.ravel())
GCV_pred=GCV.predict(x_test)
print("r2 score is ",r2_score(y_test,GCV_pred))
print("best score is ",GCV.best_score_)
print("best estimator is ",GCV.best_estimator_)
print("best param is ",GCV.best_params_)
```

```python
#Best parameter for Random forest regressor
RF=RandomForestRegressor(bootstrap=False, max_features='sqrt', n_estimators=30)
RF.fit(x_train,y_train.values.ravel())
RF_pred=RF.predict(x_test)
print("r2 score is ",r2_score(y_test,RF_pred))
print("mean squared error is ",mean_squared_error(y_test,RF_pred))
print("mean absolute error is ",mean_absolute_error(y_test,RF_pred))
```

# Scaling

```python
Scaled_clas_X=sc.fit_transform(X_clas)
```

# Best Model

```python
#Finding best model
best_rstate_clas=0
best_accu_rate=0
models=[LogisticRegression(max_iter=201), AdaBoostClassifier(), 
       GradientBoostingClassifier(), DecisionTreeClassifier(), SVC(), KNeighborsClassifier()]
for r in range(1,86):
    x_train_clas, x_test_clas, y_train_clas, y_test_clas=train_test_split(Scaled_clas_X,Y_clas,test_size=0.33,random_state=r)
    for model in models:
        model.fit(x_train_clas,y_train_clas.values.ravel())
        model_pred=model.predict(x_test_clas)
        accu_score=accuracy_score(y_test_clas,model_pred)
        if accu_score>best_accu_rate:
            best_accu_rate=accu_score
            best_rstate_clas=r
            best_model=model
            print(f"best accuracy : {best_accu_rate*100} found on the random state {best_rstate_clas}, best model is {best_model}")
print(f"best accuracy : {best_accu_rate*100} found on the random state {best_rstate_clas}, best model is {best_model}")
```

# Hyperparameter Tuning

```python
#Grid Search CV for Gradient Boost Classifier
x_train_clas, x_test_clas, y_train_clas, y_test_clas=train_test_split(Scaled_clas_X,Y_clas,test_size=0.33,random_state=85)
parameters_clas = {
    "loss":["deviance"],
    "learning_rate": [0.01, 0.1],
    "max_depth":[3,5],
    "max_features":["log2"],
    "criterion": ["friedman_mse"],
    "subsample":[0.5, 1.0],
    "n_estimators":[10]
    }
GCV_clas= GridSearchCV(GradientBoostingClassifier(), parameters_clas,scoring='accuracy',cv=5, n_jobs=-1)
GCV_clas.fit(x_train_clas, y_train_clas)
GCV_pred_clas=GCV.predict(x_test_clas)
print(GCV_clas.best_params_)
print(GCV_clas.best_estimator_)
print(GCV_clas.best_score_)
```

```python
#Finding best CV core and accuracy score
GBC=GradientBoostingClassifier(max_features='log2', n_estimators=10)
GBC.fit(x_train_clas, y_train_clas)
pred_GBC=GBC.predict(x_test_clas)
lss=accuracy_score(y_test_clas,pred_GBC)
for j in range(2,20):
    lssscore=cross_val_score(GBC,Scaled_clas_X,Y_clas,cv=j)
    lsc=lssscore.mean()
    print("At CV : ",j)
    print("cross validation score is :",lsc*100)
    print("accuracy_score is :",lss*100)
    print('\n')
```

```python
#Classification Report
from sklearn.metrics import classification_report
print(classification_report(y_test_clas,pred_GBC))
```

```python
#Final CV score and accuracy score
lssscore_selected=cross_val_score(GBC,Scaled_clas_X,Y_clas,cv=18).mean()
print("The CV score is",lssscore_selected*100," \nThe accuracy score is",lss*100)
```
