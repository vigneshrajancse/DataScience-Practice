# Scaling

```python
#Assigning input and target data
X=new_df.drop(['class'],axis=1)
Y=new_df[['class']]
```

```python
#Scaling data using Standard Scaler
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
ScaledX=sc.fit_transform(X)
```

```python
#importing imporant libraries for models
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier, RandomForestClassifier, BaggingClassifier, ExtraTreesClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

#importing train test split library
from sklearn.model_selection import cross_val_score, train_test_split

#importing error metrics
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, roc_curve
from sklearn.model_selection import GridSearchCV, cross_val_score
```

```python
#Finding best model and random state
best_rstate_clas=0
best_accu_rate=0
models=[RandomForestClassifier(), LogisticRegression(), AdaBoostClassifier(), 
       GradientBoostingClassifier(), DecisionTreeClassifier(), SVC(), KNeighborsClassifier(), GaussianNB(), BaggingClassifier(), ExtraTreesClassifier()]
for r in range(1,46):
    x_train, x_test, y_train, y_test=train_test_split(ScaledX,Y,test_size=0.33,random_state=r, stratify=Y)
    for model in models:
        model.fit(x_train,y_train)
        pred_test=model.predict(x_test)
        pred_train=model.predict(x_train)
        accu_score=accuracy_score(y_test,pred_test)
        if accu_score>best_accu_rate: 
            if round(accuracy_score(y_train,pred_train)*100,1)==round(accuracy_score(y_test,pred_test)*100,1):
                print("best random state",r)
                print("At random state",r)
                print("Training score :",accuracy_score(y_train,pred_train)*100)
                print("Testing score :",accuracy_score(y_test,pred_test)*100)
                print("Model : ", model)
                best_rstate_clas=r
                best_accu_rate=accu_score
```

Observation :

AdaBoostClassifier is the best model with random state of 45.

```python
x_train, x_test, y_train, y_test=train_test_split(ScaledX,Y,test_size=0.33,random_state=45)

param_grid_ABC = {'n_estimators':[500,1000,2000],'learning_rate':[.001,0.01,.1]}
gridsearch_ABC=GridSearchCV(AdaBoostClassifier(), param_grid_ABC, refit=True)
gridsearch_ABC.fit(x_train,y_train)
pred_ABC=gridsearch_ABC.predict(x_test)
ABC_best_parm=gridsearch_ABC.best_params_
print("Best Param ",gridsearch_ABC.best_params_)
print("Best estimator ",gridsearch_ABC.best_estimator_)
print("Best score ",gridsearch_ABC.best_score_)
```

Observation : 

AdaBoostClassifier(learning_rate=0.01, n_estimators=500) is the best parameter.

```python
AB=AdaBoostClassifier( learning_rate=0.01, n_estimators=500)
AB.fit(x_train,y_train)
pred_AB=AB.predict(x_test)
lss=accuracy_score(y_test,pred_AB)
for j in range(2,6):
    lssscore=cross_val_score(AB,ScaledX,Y,cv=j)
    lsc=lssscore.mean()
    print("At CV : ",j)
    print("cross validation score is :",lsc*100)
    print("accuracy_score is :",lss*100)
    print('\n')
```

Observation :

CV 4 is the best value.

```python
#Classification Report
print(classification_report(y_test,pred_AB))
```

```python
#Final CV score and accuracy score
lssscore_selected=cross_val_score(AB,ScaledX,Y,cv=4).mean()
print("The CV score is",lssscore_selected," \nThe accuracy score is",lss)
```

Observation:

Final score of AdaBoostClassifier are:

                    *  The CV score is 0.9238  

                    *  The accuracy score is 0.9303030303030303
