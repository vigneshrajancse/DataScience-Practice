```python
embark=pd.get_dummies(df.Embarked,prefix='Embarked')
embark.head()
```

```python
df=df.join(embark)
df.drop(['Embarked'],axis=1,inplace=True)
```

```python
df
```

```python
df.Sex=df.Sex.map({'male':0,'female':1})
```

```python
df.head()
```

```python
Y=df.Survived.copy()
```

```python
Y
```

```python
X=df.drop(['Survived'],axis=1)
X
```

```python
X.drop(['Name','Ticket'],axis=1,inplace=True)
```

```python
X.drop(['Cabin','PassengerId'], axis = 1, inplace = True)
```
